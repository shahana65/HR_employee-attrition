```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df["Attrition"])
```

```python
df["Attrition_num"]=df["Attrition"].map({"Yes":1,"No":0})
```

```python
print("Average age:",np.mean(df["Age"]))
```

```python
print("Average Monthly Income:",np.mean(df["MonthlyIncome"]))
```

```python
dept_attrition=df.groupby("Department")["Attrition_num"].mean()
print(dept_attrition)
```

```python
jobrole_income=df.groupby("JobRole")["MonthlyIncome"].mean()
print(jobrole_income)
```

```python
sns.countplot(x="Attrition",data=df)
plt.show()
```

```python
plt.hist(df["Age"],bins=10)

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="Attrition",y="MonthlyIncome",data=df)
plt.title("Salary vs Attrition")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=False)
plt.title("correlation heatmap")
plt.show()
```
